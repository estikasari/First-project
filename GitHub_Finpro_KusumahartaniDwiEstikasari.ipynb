Metadata:

**MRN**
Medical record number which is a unique identifier for an individual patient.

**LOG_ID**
Each row of this table contains a unique LOG_ID, which represents a single patient encounter. It is possible for this table to have duplicate MRNs, indicating that a single patient has come in for multiple surgeries.

**BIRTH_DATE, HEIGHT, WEIGHT, SEX**
BIRTH_DATE gives the patient age in years, HEIGHT gives the patient height in ft and inches, WEIGHT gives the patient weight in grams, and sex gives the genotypical sex of the patient.

**ASA_RATING_C, ASA_RATING**
ASA_RATING_C gives the ASA Physical Status Class 1-6 for each patient and the ASA_RATING provides the name for each class.
 **HOSP_ADMSN_TIME, HOSP_DISCH_TIME, LOS**

HOSP_ADMSN_TIME provides the date and time the patient was admitted to the hospital, HOSP_DISCH_TIME provides the date and time the patient was discharged from the hospital, and LOS provides the length of stay in the hospital in days.

**DISCH_DISP_C, DISCH_DISP**
DISCH_DISP_C represents a class 1-100 representing a patients discharge disposition. These classes correlate with the name for each discharge disposition which can be found in DISCH_DISP.

**SURGERY_DATE**
SURGERY_DATE provides the date on which the patient underwent a surgical procedure.

**PRIMARY_PROCEDURE_NM**
PRIMARY_PROCEDURE_NM indicates the name of the surgical procedure a patient is scheduled to undergo.

**IN_OR_DTTM, OUT_OR_DTTM**
IN_OR_DTTM provides the date and time the patient went into the OR and the procedure was started, and OUT_OR_DTTM provides the date and time the patient left the OR when the procedure was completed.

**AN_START_DATETIME, AN_STOP_DATETIME**
AN_START_DATETIME represents the date and time at which anesthesia was started and AN_STOP_DATETIME represents the date and time at which anesthesia was stopped.  

**ICU_ADMIN**
ICU_ADMIN displays whether or not a patient was admitted to the ICU during their visit.

**PRIMARY_ANES_TYPE_NM**
PRIMARY_ANES_TYPE_NM is the primary type of anesthesia a patient received. (ex: general, MAC, etc)

**PATIENT_CLASS_GROUP, PATIENT_CLASS_NM**
PATIENT_CLASS_GROUP specifies whether an encounter was inpatient or outpatient and PATIENT_CLASS_NM provides more detail.

# **Data pre-cleaning**




```python
# Melakukan penambahan variabel Length of Stay after Surgery (LOS_AS)
df['SURGERY_DATE'] = pd.to_datetime(df['SURGERY_DATE'])
df['HOSP_DISCH_TIME'] = pd.to_datetime(df['HOSP_DISCH_TIME'])

# Membuat kolom baru Length of Stay after Surgery (LOS_AS) dalam durasi hari
df['LOS_AS'] = (df['HOSP_DISCH_TIME'] - df['SURGERY_DATE']).dt.days
```

```python
# Melihat jumlah nilai 0 pada variabel 'LOS_AS'
nol_los_as = (df['LOS_AS'] == 0).sum()
nol_los_as
```

```python
# Mengganti nilai 0 pada variabel 'LOS_AS' menjadi 0.1
df['LOS_AS'] = df['LOS_AS'].replace(0, 0.1)
df['LOS_AS'] = df['LOS_AS'].astype(float)
```

```python
# Menambahkan variabel 'LOS_AS_CAT' yaitu kategorisasi dari variabel 'LOS_AS'
bins = [-1, 1, 3, 7, 10, 14, 21, 28, 56, 112, 224, 365]
labels = ['0-1', '2-3', '4-7', '8-10', '11-14', '15-21', '22-28', '29-56', '57-112', '113-224', '225-365']

df['LOS_AS_CAT'] = pd.cut(df['LOS_AS'], bins=bins, labels=labels)
```

```python
# Menambahkan variabel Length of Stay in Operating Room (LOR)
df['IN_OR_DTTM'] = pd.to_datetime(df['IN_OR_DTTM'])
df['OUT_OR_DTTM'] = pd.to_datetime(df['OUT_OR_DTTM'])

# Membuat kolom baru 'LOR' dalam durasi menit
df['LOR'] = (df['OUT_OR_DTTM'] - df['IN_OR_DTTM']).astype('timedelta64[m]')
```

```python
# Melakukan ekstraksi variabel 'SURGERY_DATE' menjadi nilai bulan dan tahun
df['SURGERY_MONTH'] = df['SURGERY_DATE'].dt.month
df['SURGERY_YEAR'] = df['SURGERY_DATE'].dt.year
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Fungsi untuk membuat model Keras
def create_keras_model(optimizer='adam', activation='relu', neurons=64):
    model_keras = Sequential()
    model_keras.add(Dense(neurons, input_dim=X_train.shape[1], activation=activation))
    model_keras.add(Dense(32, activation=activation))
    model_keras.add(Dense(1, activation='linear'))
    model_keras.compile(optimizer=optimizer, loss='mean_squared_error')
    return model_keras

# Fungsi untuk menghitung MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Membuat daftar parameter yang akan dicoba
param_grid_keras = {
    'optimizer': ['adam', 'sgd'],
    'activation': ['relu', 'tanh'],
    'neurons': [32, 64, 128]
}

best_model_keras = None
best_params_keras = None
best_rmse_test_keras = float('inf')
best_rmse_train_keras = float('inf')
best_r2_test_keras = float('-inf')
best_r2_train_keras = float('-inf')
best_mae_test_keras = float('inf')
best_mae_train_keras = float('inf')

# Melakukan iterasi melalui kombinasi parameter
for params in ParameterGrid(param_grid_keras):
    model_keras = create_keras_model(**params)

    # Melatih model
    model_keras.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_test, Y_test), verbose=0)

    # Melakukan prediksi pada data uji dan data latih
    Y_pred_test_keras = model_keras.predict(X_test)
    Y_pred_train_keras = model_keras.predict(X_train)

    # Evaluasi model
    rmse_test_keras = np.sqrt(mean_squared_error(Y_test, Y_pred_test_keras))
    rmse_train_keras = np.sqrt(mean_squared_error(Y_train, Y_pred_train_keras))

    r2_test_keras = r2_score(Y_test, Y_pred_test_keras)
    r2_train_keras = r2_score(Y_train, Y_pred_train_keras)

    mae_test_keras = mean_absolute_error(Y_test, Y_pred_test_keras)
    mae_train_keras = mean_absolute_error(Y_train, Y_pred_train_keras)

    mape_test_keras = mean_absolute_percentage_error(Y_test, Y_pred_test_keras)
    mape_train_keras = mean_absolute_percentage_error(Y_train, Y_pred_train_keras)

    # Menyimpan model terbaik
    if rmse_test_keras < best_rmse_test_keras:
        best_rmse_test_keras = rmse_test_keras
        best_rmse_train_keras = rmse_train_keras
        best_model_keras = model_keras
        best_params_keras = params
        best_r2_test_keras = r2_test_keras
        best_r2_train_keras = r2_train_keras
        best_mae_test_keras = mae_test_keras
        best_mae_train_keras = mae_train_keras

# Menampilkan hasil evaluasi model terbaik
print("Best Parameters:", best_params_keras)
print(f'Root Mean Squared Error (RMSE) on Test Data: {best_rmse_test_keras}')
print(f'R-squared (R2) on Test Data: {best_r2_test_keras}')
print(f'Mean Absolute Error (MAE) on Test Data: {best_mae_test_keras}')
print(f'Mean Absolute Percentage Error (MAPE) on Test Data: {mape_test_keras}')

print(f'Root Mean Squared Error (RMSE) on Training Data: {best_rmse_train_keras}')
print(f'R-squared (R2) on Training Data: {best_r2_train_keras}')
print(f'Mean Absolute Error (MAE) on Training Data: {best_mae_train_keras}')
print(f'Mean Absolute Percentage Error (MAPE) on Training Data: {mape_train_keras}')
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Menggunakan OneVsRestClassifier untuk tugas klasifikasi multiclass
classifier_rfc = OneVsRestClassifier(grid_search_rfc.best_estimator_)

# Binarize label
n_classes_rfc = len(np.unique(Y_class_test))
y_bin_rfc = label_binarize(Y_class_test, classes=np.unique(Y_class_test))

# Latih model dengan data oversampled dan memperoleh probabilitas
y_score_rfc = classifier_rfc.fit(X_class_train, Y_class_train).predict_proba(X_class_test)

# Menghitung kurva ROC untuk setiap kelas
fpr_rfc = dict()
tpr_rfc = dict()
roc_auc_rfc = dict()
for i in range(n_classes_rfc):
    fpr_rfc[i], tpr_rfc[i], _ = roc_curve(y_bin_rfc[:, i], y_score_rfc[:, i])
    roc_auc_rfc[i] = auc(fpr_rfc[i], tpr_rfc[i])

# Plotting kurva ROC untuk setiap kelas
plt.figure(figsize=(8, 6))
for i in range(n_classes_rfc):
    plt.plot(fpr_rfc[i], tpr_rfc[i], label=f'Class {i} (AUC = {roc_auc_rfc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Multiclass')
plt.legend(loc='lower right')
plt.show()
```
